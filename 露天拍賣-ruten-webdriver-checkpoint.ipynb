```python
import requests as rq
import html5lib
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup 
import re

ruten={}
print("露天拍賣")

for pp in range(1,6):
    page="頁次-"+str(pp)
    print(page)

    url = "http://find.ruten.com.tw/search/s000.php?enc=u&t=0&f=0&f1=&f2=&searchfrom=detailf&k=%E8%80%81%E6%A5%8A&c=0024&fy=0&fz=0&p1=&p2=&h=0&f1x=&f1y=&f2x=&f2y=&p="+str(pp) 

    driver=webdriver.Chrome()  #调用本地的火狐浏览器，Chrom 甚至 Ie 也可以的
    driver.get(url) #请求页面，会打开一个浏览器窗口
    html_text=driver.page_source
    driver.quit()
    #print (html_text)
    soup=BeautifulSoup(html_text, "lxml")
    
    for div in soup.findAll('div',{"class":"media-body"}):   
        for span in div.findAll('span',{"class":"item-name-text"}):
            ruten['商品']='商品:'+span.text+' '
            print(ruten['商品'], end='')
        for price in div.findAll('strong',{"class":"rt-text-price"}):
            ruten['單價']=' ,單價:'+price.text.replace(" ","").replace("\n","")+' '
            print(ruten['單價'])
    
    
print("FINISH")
```
