```python
import requests as rq
from selenium import webdriver
from bs4 import BeautifulSoup #导入包
import re
import time

shopee={}
print("蝦皮拍賣")


for pp in range(0,8):    
    url = "https://shopee.tw/search/?is_official_shop=false&keyword=老楊&page="+str(pp) 
    driver=webdriver.Chrome()  #调用本地的火狐浏览器，Chrom 甚至 Ie 也可以的
    #time.sleep(60)
    driver.get(url) #请求页面，会打开一个浏览器窗口
    html_text=driver.page_source
    driver.quit()
    soup=BeautifulSoup(html_text, "lxml")

    page="頁次-"+str(pp)
    print(page)
    
    for div in soup.findAll('div',{"class":"shopee-item-card__lower-padding"}): 
        #print(div)
        for divname in div.findAll('div',{"class":"shopee-item-card__text-name"}):
            shopee['商品']='商品:'+divname.get_text()+' '
            print(shopee['商品'], end='')
        for divprice in div.findAll('div',{"class":"shopee-item-card__current-price"}):
            shopee['單價']=' ,單價:'+divprice.get_text()+' '
            print(shopee['單價'])

print("finish")
```
