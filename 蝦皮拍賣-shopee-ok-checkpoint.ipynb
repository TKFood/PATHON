```python
import requests as rq
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup #导入包
import re
import time

shopee={}
print("蝦皮拍賣")


for pp in range(0,9):    
    url = "https://shopee.tw/search/?keyword=%E8%80%81%E6%A5%8A&order=asc&page=" +str(pp)+"&sortBy=price"
    driver=webdriver.Chrome()  #调用本地的火狐浏览器，Chrom 甚至 Ie 也可以的
    #time.sleep(60)
    driver.get(url) #请求页面，会打开一个浏览器窗口   
    
    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "search-result"))
        )

        html_text=driver.page_source
        soup=BeautifulSoup(html_text, "lxml")

        print("")
        page="頁次-"+str(pp)
        print(page)

        for div in soup.findAll('div',{"class":"shopee-item-card__lower-padding"}): 
            #print(div)
            for divname in div.findAll('div',{"class":"shopee-item-card__text-name"}):
                shopee['商品']='商品:'+divname.get_text()+' '
                print(shopee['商品'], end='')
            for divprice in div.findAll('div',{"class":"shopee-item-card__current-price"}):
                shopee['單價']=' ,單價:'+divprice.get_text()+' '
                print(shopee['單價'])
            
    finally:
        driver.quit()
    
    

print("finish")
```
