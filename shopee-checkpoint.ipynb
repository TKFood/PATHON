```python
import requests as rq
from selenium import webdriver
from bs4 import BeautifulSoup #导入包
import re
import time

FRIADY={}
print("蝦皮拍賣")


for pp in range(0,9):    
    url = "https://shopee.tw/search/?is_official_shop=false&keyword=%E8%80%81%E6%A5%8A&page="+str(pp) 
    driver=webdriver.Chrome()  #调用本地的火狐浏览器，Chrom 甚至 Ie 也可以的
    #time.sleep(60)
    driver.get(url) #请求页面，会打开一个浏览器窗口
    html_text=driver.page_source
    driver.quit()
    soup=BeautifulSoup(html_text, "lxml")

    page="頁次-"+str(pp)
    print(page)
    
    for div in soup.findAll('div',{"class":"shopee-search-result-view__item-card"}):
        for linkA in div.findAll('a',{"class":"shopee-item-card--link"}):        
            #print("https://shopee.tw/"+linkA['href'])
            childurl="https://shopee.tw/"+linkA.get('href')
            childresponse = rq.get(childurl) # 用 requests 的 get 方法把網頁抓下來
            childhtml_doc = childresponse.text # text 屬性就是 html 檔案
            childsoup = BeautifulSoup(childhtml_doc, "lxml") # 指定 lxml 作為解析器
            
            for childdiv in childsoup.find_all("div",{"class":"shopee-product-info__header__left-section"}): 
                for childtitle in childdiv.find_all('div',{"class":"shopee-product-info__header__title"}):
                    for h1 in childtitle.find_all('h1',{"class":"shopee-product-info__header__text"}):
                        print(h1.text)
                    
        
        
print("FINISH")
```
