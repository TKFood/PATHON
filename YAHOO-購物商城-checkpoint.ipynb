```python
import requests as rq 
import html5lib
from urllib.request import urlopen
from bs4 import BeautifulSoup 

YAHOOdata={}

print("YAHOO")

url = "https://tw.buy.yahoo.com/?sub=97&sort=-ptime&pg=1" 
html=urlopen(url)
bsobj=BeautifulSoup(html.read(),"html5lib")

response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
html_doc = response.text # text 屬性就是 html 檔案
soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器

for div in soup.findAll('div',{"class":"srp-pdtitle"}):
    for linkA in div.findAll('a'):
        
        childurl=linkA.get('href')
        #print(childurl)
        childresponse = rq.get(childurl) # 用 requests 的 get 方法把網頁抓下來
        childhtml_doc = childresponse.text # text 屬性就是 html 檔案
        childsoup = BeautifulSoup(childhtml_doc, "lxml") # 指定 lxml 作為解析器
        
        for childtitle in childsoup.find_all("div",{"class":"title"}):
            for childname in childtitle.findAll('h1'):
                YAHOOdata['商品']='商品:'+childname.text
                print(YAHOOdata['商品'], end='')
        
        for childdiv in childsoup.find_all("div",{"class":"priceinfo"}):
            for childprice in childdiv.find_all("span",{"class":"price"}):
                YAHOOdata['單價']=' ,單價:'+childprice.text
                print(YAHOOdata['單價'])


print("FINISH")
print("FINISH")
```
