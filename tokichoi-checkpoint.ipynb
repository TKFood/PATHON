```python
import requests as rq
import html5lib
from urllib.request import urlopen
from bs4 import BeautifulSoup 
import re

url = "https://www.tokichoi.com.tw/Product/Category/pdmc0174/tokichoi?utm_source=AsiaPac_google&utm_medium=ROI&utm_campaign=BRAND" 
response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
html_doc = response.text # text 屬性就是 html 檔案
soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器


for div in soup.findAll('div',{"id":"product"}):
    for name in div.findAll('a',{"class":"photo CategoryProductButtonLink"}):        
        for prize in div.findAll('p',{"class":"prize"}):
            print(name["title"])
            print('單價:'+prize.text)

print("FINISH")
```
