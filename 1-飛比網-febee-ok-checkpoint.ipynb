```python
import requests
import html5lib
from urllib.request import urlopen
from bs4 import BeautifulSoup 

print("feebee")

for pp in range(0,43):
    url="http://feebee.com.tw/all/%E8%80%81%E6%A5%8A/?sort=p&mode=l&ptab=0&page="+str(pp)
    html=urlopen(url)
    bsobj=BeautifulSoup(html.read(),"html5lib")

    page="頁次-"+str(pp)

    
    
    for childli in bsobj.find_all("li"):         
        for childdiv in childli.find_all("span"): 
            for name in childdiv.find_all("h2",{"class":"large"}): 
                print(page, end='')
                nameF=name.get_text().replace(" ","").replace("\n","")
                print(","+nameF, end='')
            for price in childdiv.find_all("div",{"class":"price ellipsis xlarge"}):
                print(","+price.get_text() , end='')
            for source in childdiv.find_all("",{"class":"shop"}):
                sourceF=source.get_text().replace(" ","").replace("\n","")
                print (","+sourceF )
                

print("FINISH") 
```
